```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('img/img/dog.jpg')
type(img)
```

#### BGR format

```python
plt.figure(figsize=(12,8))
plt.imshow(img)
plt.title('BGR Format')
plt.show()
```

#### RGB format

```python
rgb_img=img.copy()
rgb_img=cv2.cvtColor(rgb_img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(rgb_img)
plt.title('RGB Format')
plt.show()
```

### THRESHOLD

```python
gray_img=rgb_img.copy()
gray_img=cv2.cvtColor(gray_img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(12,8))
plt.imshow(gray_img, cmap='gray')
plt.title('GRAY Format')
plt.show()
```

```python
retval, dst= cv2.threshold(gray_img,200,255,cv2.THRESH_BINARY)
plt.figure(figsize=(12,8))
plt.imshow(dst, cmap='gray')
```

```python
retval, dst= cv2.threshold(gray_img,200,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize=(12,8))
plt.imshow(dst, cmap='gray')
```

```python
retval, dst= cv2.threshold(gray_img,200,255,cv2.THRESH_TOZERO)
plt.figure(figsize=(12,8))
plt.imshow(dst, cmap='gray')
```

```python
retval, dst= cv2.threshold(gray_img,200,255,cv2.THRESH_TOZERO_INV)
plt.figure(figsize=(12,8))
plt.imshow(dst, cmap='gray')
```
