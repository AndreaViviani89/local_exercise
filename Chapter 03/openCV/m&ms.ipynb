```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('Computer Vision Course Resources/img/img/m.jpg')
hsv_img = cv2.imread('Computer Vision Course Resources/img/img/hsv.png')
```

```python
img
```

```python
hsv_img = cv2.cvtColor(hsv_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(18, 12))
plt.imshow(hsv_img)
```

### Isolate the blue

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower_blue = np.array([85, 50, 50])
upper_blue = np.array([130, 255, 255])
```

```python
mask_blue = cv2.inRange(img_hsv, lower_blue, upper_blue)
res_blue = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_blue)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(res_blue)
```

### Isolate the green

```python
lower_green = np.array([40, 50, 50])
upper_green = np.array([70, 255, 255])

mask_green = cv2.inRange(img_hsv, lower_green, upper_green)
res_green = cv2.bitwise_and(img_rgb, img_rgb, mask= mask_green)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(res_green)
```

### Showing only blue and green

```python
mask_gb = cv2.bitwise_or(mask_blue, mask_green)
img_gb = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_gb)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(mask_gb)
plt.imshow(img_gb)
```

### Showing blue, green and red

```python
lower_red_1 = np.array([0,50,50])
upper_red_1 = np.array([10,255,255])
lower_red_2 = np.array([170,50,50])
upper_red_2 = np.array([180,255,255])
```

```python
mask_red_1 = cv2.inRange(img_hsv, lower_red_1, upper_red_1)
mask_red_2 = cv2.inRange(img_hsv, lower_red_2, upper_red_2)
mask_red = mask_red_1 + mask_red_2
mask_rgb = cv2.bitwise_or(mask_red, mask_gb)
res_rgb = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_rgb)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(res_rgb)
```
